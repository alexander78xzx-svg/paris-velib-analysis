```python
import pandas as pd
```

```python
df = pd.read_csv(".\\datasets\\may-june-2024.csv")
```

```python
df.head()
```

```python
df['Departure_Time'] = pd.to_datetime(df['Departure_Time'])
df = df.dropna(subset=['Departure_Station_Name', 'Departure_Time']) 
df['Duration'] = pd.to_timedelta(df['Duration']) 
df['Duration_Minutes'] = df['Duration'].dt.total_seconds() / 60
```

```python

columns_to_keep = [
    'Departure_Time', 
    'Departure_Station_Name', 
    'Electric', 
    'Duration',
    'Duration_Minutes',
    'Departure_Latitude',
    'Departure_Longitude'
]
df_clean = df[columns_to_keep].copy()
```

```python
df_clean['Hour'] = df_clean['Departure_Time'].dt.hour
df_clean['Weekday'] = df_clean['Departure_Time'].dt.day_name()
df_clean['Is_Weekend'] = df_clean['Weekday'].isin(['Saturday', 'Sunday'])
```

```python
df_clean.to_csv('velib_clean_for_tableau.csv', index=False, encoding='utf-8')
```
